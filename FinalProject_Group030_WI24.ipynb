## Background and Prior Work


The United States has a long history of using the death penalty as a punishment for certain crimes, including homicide. However, attitudes toward capital punishment and its prevalence have varied over time. Central to the debate on the death penalty's effect on homicide rates is the theory of deterrence, which posits that the threat of severe punishment, such as execution, can deter individuals from committing crimes. 

The Naci Mocan study, highlighted in the Washington post’s "Murderous Pardons?" article, investigates the link between death penalty execution and homicides rate in the U.S. from 1977 to 1997. The study examined 6143 death sentences between 1977-1997, and included 1050 observations (one per state for 21 years) in their dataset. The study found that each execution correlated with five fewer homicides, and removal from death row resulted in one more homicide. However, The study has been criticized by two criminologists, Richard Berk and Jeffrey Fagan. There might be flaws in the study’s methodology as of the skewed distributions of execution figures across states and years. While excluding Texas, the deterrent effect of executions disappear. In addition, Fagan points out that there might be flaws on Mocan’s deterrence measure, indicating that the study may have inappropriate generalization upon the selected observations. <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) 

Many people have argued that abolishing the death penalty leads to higher crime rates, but studies in the USA do not back this up. In 2004 in the USA, the average murder rate for states that used the death penalty was 5.71 per 100,000 of the population, against 4.02 per 100,000 in states that did not use it.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) The death penalty has been shown to have a brutalizing effect on society. State sanctioned killing endorses the use of force and continues the cycle of violence. 

It is important to question the validity in the theories that support the death penalty as this legal reform comes at a very high cost. The death penalty is irrevocable; however, the justice system is not immune to errors: corruption, false confessions, misidentifications, systematic inequalities have all played a part in the wrongful conviction of innocent people. There have been many instances in the past where innocent lives have been tragically taken away only to be proven innocent later on. Studies have shown that 4 percent of people on death row are innocent –  in California that is 26 lives.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Furthermore, in California, while African Americans are only 6 percent of the population, they make up 34 percent.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) A global movement to abolish the death penalty in support of basic human rights has been gaining momentum across the globe, and it is time for California to join this movement and eliminate these outdated and flawed practices.

Now that we have dataframes for each of the states of interest, we want to fill in the missing years so each state has data ranging from 1980-2014

```python
#Find number of executions by year ranging 1980-2023 for Illinois
df_Illinois_execution['Year'] = df_Illinois_execution['Execution Date'].dt.year
Illinois_executions_by_year = df_Illinois_execution.groupby(['Year','State']).size().reset_index(name='Count')
all_years = range(1980, 2024)
all_years_df = pd.DataFrame({'Year': all_years})
Illinois_executions_by_year = pd.merge(all_years_df, Illinois_executions_by_year, on = 'Year', how='left')
Illinois_executions_by_year['Count'] = Illinois_executions_by_year['Count'].fillna(0)
Illinois_executions_by_year['State'] = Illinois_executions_by_year['State'].fillna('Illinois')

#Find number of executions by year ranging 1980-2023 for Texas
df_texas_execution['Year'] = df_texas_execution['Execution Date'].dt.year
texas_executions_by_year = df_texas_execution.groupby(['Year','State']).size().reset_index(name='Count')
all_years = range(1980, 2024)
all_years_df = pd.DataFrame({'Year': all_years})
texas_executions_by_year = pd.merge(all_years_df, texas_executions_by_year, on = 'Year', how='left')
texas_executions_by_year['Count'] = texas_executions_by_year['Count'].fillna(0)
texas_executions_by_year['State'] = texas_executions_by_year['State'].fillna('Texas')

#Find number of executions by year ranging 1980-2023 for California
df_california_execution['Year'] = df_california_execution['Execution Date'].dt.year
california_executions_by_year=df_california_execution.groupby(['Year','State']).size().reset_index(name='Count')
all_years = range(1980, 2024)
all_years_df = pd.DataFrame({'Year': all_years})
california_executions_by_year = pd.merge(all_years_df, california_executions_by_year, on = 'Year', how='left')
california_executions_by_year['Count'] = california_executions_by_year['Count'].fillna(0)
california_executions_by_year['State'] = california_executions_by_year['State'].fillna('California')
```

### Illinois Homicide Rate Analysis

Let's see if there is a significant difference in murder rate before 2011, which was when Illinois repealed the death penalty, and after.

```python
# Calculate the murder rate based on population
murder_rate = []

for i in range(len(IL_df)):
    rate = IL_df.loc[i, 'murder'] / IL_df.loc[i, 'population'] * 100000
    murder_rate.append(rate)
```

```python
# Add murder rate to the dataset
IL_df['murder_rate'] = murder_rate
IL_df.head()
```

```python
MR_df = IL_df[['year', 'murder_rate']].groupby('year').mean().reset_index()
MR_df.head()
```

```python
# Generate a lineplot to show the average murder rate between the years 2007 and 2015
plt.figure(figsize=(10,4))
sns.lineplot(x='year', y='murder_rate', data=MR_df)

plt.xlabel('Year')
plt.ylabel('Homicide Rate (per 100,000)')
plt.title('Average Homicide Rate in Illinois');
```

Our maximal points are in years 2009, 2011 and 2013 with 2013 because slightly higher than the rest; however, there isn't a big difference in murder rate before 2011 and after. Since 2011 was when Illinois abolished their death penalty, this suggests that there might not be a relationship between homicide rate and executation rate. There seems to be a trend of rising and falling of homocide rates. Since 2013, the homicide rate has continued to decrease.

```python
group_1 = MR_df[MR_df['year'].isin(range(2007,2011))]
group_2 = MR_df[MR_df['year'].isin(range(2012,2016))]
combined = pd.concat([group_1.assign(Period = '2007-2010'), group_2.assign(Period='2012-2015')])
plt.figure(figsize=(10, 6))
sns.boxplot(x='Period', y='murder_rate',data=combined, palette='Blues')
plt.title('Illinois Homicide Rates: 2007-2010 vs 2012-2015')
plt.xlabel('Time Period')
plt.ylabel('Homicide Rate (per 100,000)')
plt.tight_layout()
plt.show()
```

The larger boxplot for 2007-2010 with a slightly higher median compared to 2012-2015 suggests that there was greater variability in homicide rates before the abolishment of the death penalty in 2011. However, the years 2012-2015 have an outlier suggesting that there could be instances of high homicide rates post the abolishment of the death penalty in 2011.

We will perform a t-test on the murder rate from 2007-2015 which is in our dataframe MF_df. Our null hypothesis will be the average murder rate before the death penalty was repealed (2011).

```python
# Get the average murder rate from 2007-2010
before_penalty = MR_df[MR_df['year'] < 2011]
before_mean = before_penalty.mean()['murder_rate']

# Perform one-sample t-test
t_stat, p_val = stats.ttest_1samp(MR_df['murder_rate'], before_mean)

# Print results
print('T-statistic:', t_stat)
print('P-value:', p_val)
```

The p-value is higher than 0.05 therefore it is not signifcant and we do not have enough evidence to reject the null hypothesis. Thus, there wasn't much change in murder rate before and after the repeal of the death penalty.

### Texas Homicide and Execution Analysis

Since Texas did not abolish the death penalty, let's look at the relationship between homicide rate and execution rate over the time period 1980-2014. First we will normalize the homicide incident rates and execution rates. 

```python
#Merge datasets of execution count, incidents count, population count as one dataset
df_texas_yr1 = pd.merge(texas_executions_by_year, df_incident_yr, on='Year')
df_texas_yr = pd.merge(df_texas_yr1, df_population, on='Year')
df_texas_yr

#Normalize to execution rate following Execution Rate = (Number of Executions / Population) * 100,000
df_texas_yr['Execution Rate'] = (df_texas_yr['Count']/df_texas_yr[' Population']) * 100000  

#Normalize to homocide following Execution Rate = (Number of homocides / Population) * 100,000
df_texas_yr['Homocide Rate'] = (df_texas_yr['Incidents Count']/df_texas_yr[' Population']) * 100000

#Preview Dataframe
df_texas_yr.head()
```

```python
#Plot Execution and Homicide Rate
fig, homicide = plt.subplots()
homicide.plot(df_texas_yr['Year'], df_texas_yr['Homocide Rate'], color = 'purple', label='Homicide Rate')
homicide.set_ylabel('Homicide Rate (per 100,000)')
homicide.set_xlabel('Year')

execution = homicide.twinx()
execution.plot(df_texas_yr['Year'], df_texas_yr['Execution Rate'], color = 'green', label='Execution Rate')
execution.set_ylabel('Execution Rate (per 100,000)')

homicide.legend(loc='upper left')
execution.legend(loc='upper right')

plt.show()
```

The homicide rate for Texas has decreased throughout the years, while the execution rate increased and then started to decrease overall after 2000. However, it is difficult to determine the exact relationship between execution rate and homicide rates in Texas, so let's look at the correlation between execution rate and homicide rate over the time period 1980-2014

```python
#Scatterplot with line of best fit
green_color = sns.color_palette()[2]
plt.figure(figsize=(12, 6))
sns.regplot(x='Homocide Rate', y='Execution Rate', data=df_texas_yr, scatter_kws={'s': 100}, color=green_color) 
plt.xlabel('Homicide Rate (per 100,000)') 
plt.ylabel('Execution Rate (per 100,000)') 
plt.title('Texas Homicide and Execution Rates') 
plt.show()
```

After creating the above scatterplot, we can see that there seems to be negative correlation between the Execution Rate and Homicide Rate for Texas. However, this is just based on the visualization. It is necessary to calculate the correlation coefficient to see their relationship.

```python
#seeing the correlation between homocide rate and execution rate
correlation = np.corrcoef(df_texas_yr['Execution Rate'], df_texas_yr['Homocide Rate'])[0, 1]
correlation
```

The correlation coeﬀicient of -0.595 between homicide rate and execution rate in Texas from 1980 to 2014 suggests a moderate negative correlation between these variables. This indicates that as the execution rate increases, the homicide rate tends to decrease, and vice versa. While correlation does not imply causation, this finding may suggest a potential inverse relationship between capital punishment and homicide rates in Texas. Since there is a potential inverse relationship between capital punishment and homicide rate, this could provide evidence for statistically significant relationship between execution rate and homicide, possibly supporting our hypothesis that the death penalty reduced homicide rate.

Now let's do a linear regression to gain more information on the relationship between execution and homicide rates for Texas.

```python
outcome, predictors = patsy.dmatrices('Q("Homocide Rate") ~ Q("Execution Rate")', df_texas_yr)
model = sm.OLS(outcome, predictors)
res = model.fit()
print(res.summary())
```

After performing a linear regression, there is a statistically signficant relationship between the homicide rate and execution rate for the years 1980-2014 in Texas since the p-value is less than 0.01. Thus, there is a statistically significant negative correlation between homicide rates and execution rate. When there is an the execution rate increases by 1, the homicide rate decreases by 44.4047. Therefore, indicating that the homicide rate is at its highest when the execution rate is low. This supports our hypothesis as removing the dealth penalty would increase crime rate

Now let's run a one-sided t-test to determine if the last execution in California increased homicide rates. Our null hypothesis is the mean homicide rate for 2002-2005 is greater or equalt to the mean homicide rate for 2007-2010. The alternative hypothesis is that the mean homicide rates for 2002-2005 is less than the mean homicide rates for 2007-2010. 

```python
t_statistic, p_value = stats.ttest_ind(group_1, group_2, alternative='less')
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

Since the p-value is 0.997, there is not enough evidence to conclude that the mean of homicide rates for 2002-2005 is significantly less than the mean of homicide rates for 2007-2010. Therefore, the last execution in 2006 for California did not cause an increase in homicide rates for the three years following it.
